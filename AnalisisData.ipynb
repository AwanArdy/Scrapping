```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Fungsi untuk memuat data dari file CSV
def load_data(filename):
    return pd.read_csv(filename)
```

```python
# Fungsi untuk menampilkan statistik deskriptif
def descriptive_statistics(df):
    return df.describe()
```

```python
# Fungsi untuk memvisualisasikan distribusi parameter tertentu
def plot_distribution(df, parameter_description):
    plt.figure(figsize=(10, 6))
    sns.histplot(df[df['parameter_description'] == parameter_description]['value_text'].astype(float), bins=30, kde=True)
    plt.title(f'Distribution of {parameter_description}')
    plt.xlabel(parameter_description)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Fungsi untuk memvisualisasikan data parameter tertentu dari waktu ke waktu
def plot_time_series(df, area_name, parameter_description):
    df_filtered = df[(df['area_name'] == area_name) & (df['parameter_description'] == parameter_description)]
    df_filtered['datetime'] = pd.to_datetime(df_filtered['datetime'], format='%Y%m%d%H%M')
    df_filtered['value_text'] = df_filtered['value_text'].astype(float)
    
    plt.figure(figsize=(14, 7))
    sns.lineplot(x='datetime', y='value_text', data=df_filtered)
    plt.title(f'{parameter_description} Over Time for {area_name}')
    plt.xlabel('Date Time')
    plt.ylabel(parameter_description)
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Fungsi untuk memvisualisasikan box plot dari parameter tertentu berdasarkan area
def plot_boxplot(df, parameter_description):
    plt.figure(figsize=(14, 7))
    sns.boxplot(x='area_name', y='value_text', data=df[df['parameter_description'] == parameter_description].astype({'value_text': float}))
    plt.title(f'Boxplot of {parameter_description} by Area')
    plt.xlabel('Area')
    plt.ylabel(parameter_description)
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Memuat data dari CSV
filename = 'prakiraan_cuaca_sulawesi_tengah.csv'
df_weather = load_data(filename)
```

```python
# Menampilkan statistik deskriptif
print(descriptive_statistics(df_weather))
```

```python
# Visualisasi distribusi parameter tertentu
plot_distribution(df_weather, 'Temperature')
```

```python
# Visualisasi data parameter tertentu dari waktu ke waktu
plot_time_series(df_weather, 'Palu', 'Temperature')
```

```python
# Visualisasi box plot dari parameter tertentu berdasarkan area
plot_boxplot(df_weather, 'Temperature')
```
